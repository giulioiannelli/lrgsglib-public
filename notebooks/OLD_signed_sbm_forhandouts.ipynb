```python
%matplotlib inline
import sys
sys.path.append('../')
from src.LRGSG_package.LRGSG import *
```

```python
# stochastic block model parameters
n1 = 100
n2 = 100
extra_link = 1
intra_link = 40
sizes = [n1, n2]
probs = [[intra_link/n1, extra_link/n1], [extra_link/n1, intra_link/n2]]
```

```python
G = nx.stochastic_block_model(sizes, probs, seed=0)
nnodes = G.number_of_nodes()
label_list = [i for i in range(nnodes)]
node_color = ['r' if i < n1 else 'b' for i in range(nnodes)]
#
fig, ax = plt.subplots(figsize=(15, 8))
nx.draw(G, node_size=300, node_color=node_color,  alpha=0.5)
```

```python
[Sm1, dS1, VarL1, t1] = entropy(G, is_signed=False, t2=1)
t11 = .5 * (t1[1:] + t1[:-1])

fig, ax1 = plt.subplots(figsize=(12, 8))
ax2 = ax1.twinx()

ax1.plot(t1, Sm1, 'k--')



ax2.plot(t11, dS1, 'k')

ax1.set_ylabel(r'$1-S$')
ax2.set_ylabel(r'$C\log N$')
ax1.set_xlabel(r'$\tau$')

ax1.set_xscale('log')
```

```python
# stochastic block model parameters
n1 = n2 = 50
extra_link = 1
intra_link = 20
sizes = [n1, n2]
probs = [[intra_link/n1, extra_link/n1], [extra_link/n2, intra_link/n2]]
```

```python
G = nx.stochastic_block_model(sizes, probs, seed=0)
labelList = [i for i in range(G.number_of_nodes())]
plt.figure(figsize=(12, 6))
node_color= ['r' if ni < 50 else 'b' for ni in range(G.number_of_nodes())]
nx.draw(G, node_size=300, node_color=node_color, with_labels = True)
```
